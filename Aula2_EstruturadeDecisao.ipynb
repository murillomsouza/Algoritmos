# **1. O que são Estruturas de Decisão?**

As estruturas de decisão permitem que o programa tome diferentes ações com base em condições específicas. Em Python, temos:

* if: Executa um bloco de código apenas se a condição for verdadeira.
* if else: Executa um bloco de código se a condição for verdadeira, e outro bloco se a condição for falsa.
elif: Permite testar várias condições de maneira encadeada.
* if aninhado: Um if dentro de outro, para verificar condições mais complexas.
* Essas estruturas ajudam o programa a seguir diferentes caminhos dependendo dos dados de entrada.

# **Operadores Relacionais em Python:**

**== (Igual a)**

Verifica se dois valores são iguais.
Exemplo:

```
x = 5
y = 5
print(x == y)  # True
```



**!= (Diferente de)**

Verifica se dois valores são diferentes.
Exemplo:

```
x = 5
y = 3
print(x != y)  # True
```



**> (Maior que)**

Verifica se o valor da esquerda é maior que o da direita.
Exemplo:


```
x = 7
y = 3
print(x > y)  # True
```



**< (Menor que)**

Verifica se o valor da esquerda é menor que o da direita.
Exemplo:

```
x = 2
y = 5
print(x < y)  # True
```




**>= (Maior ou igual a)**

Verifica se o valor da esquerda é maior ou igual ao da direita.
Exemplo:


```
x = 5
y = 5
print(x >= y)  # True
```




**<= (Menor ou igual a)**

Verifica se o valor da esquerda é menor ou igual ao da direita.
Exemplo:


```
x = 3
y = 7
print(x <= y)  # True
```




# **Resumindo:**
***
```
==  : Verifica se dois valores são iguais.
!=  : Verifica se dois valores são diferentes.
>   : Verifica se o valor à esquerda é maior.
<   : Verifica se o valor à esquerda é menor.
>=  : Verifica se o valor à esquerda é maior ou igual.
<=  : Verifica se o valor à esquerda é menor ou igual.
```

# **2. Estrutura if**

O comando if é a estrutura mais simples de decisão. Ele executa um bloco de código somente se a condição fornecida for verdadeira.

Sintaxe:

`if condição:`
    # código a ser executado se a condição for verdadeira


**Exemplo 1:**

Vamos verificar se o valor de x é maior que 5

````python
x = 10
if x > 5:
  print("x é maior que 5")
````

**Exemplo 2:**

Vamos verificar se o número fornecido é positivo.

````python
numero = int(input("Digite um número: "))

if numero > 0:
    print("O número é positivo.")
````

**Explicação:** O código verifica se o número fornecido é maior que zero. Se for, ele imprime "O número é positivo."

# **3. Estrutura if else**

O *if else* é uma extensão do if. Ele permite que você execute uma ação se a condição for verdadeira e outra ação se a condição for falsa.

Sintaxe:

```
if condição:
    # código a ser executado se a condição for verdadeira
else:
    # código a ser executado se a condição for falsa
```




**Exemplo 1:**

VAmos verificar se o valor de x é maior ou menor ou igual a 5

````python
x = 3
if x > 5:
    print("x é maior que 5")
else:
    print("x é menor ou igual a 5")
````

**Exemplo 2:**

Agora, vamos verificar se o número fornecido é positivo ou negativo.

````python
numero = int(input("Digite um número: "))

if numero > 0:
    print("O número é positivo.")
else:
    print("O número é negativo ou zero.")
````

**Explicação**:

Se o número for maior que zero, o programa imprime "O número é positivo". Caso contrário, imprime "O número é negativo ou zero".

# **4. Estrutura elif (else if)**

O comando elif (abreviação de "else if") é útil quando temos várias condições a serem verificadas, uma após a outra. Ele permite verificar várias possibilidades sem precisar de múltiplos if encadeados.

```
if condição1:
    # código para a primeira condição
elif condição2:
    # código para a segunda condição
else:
    # código para quando nenhuma das condições for verdadeira
```




**Exemplo 1:**

````python
x = 7
if x > 10:
    print("x é maior que 10")
elif x > 5:
    print("x é maior que 5, mas menor ou igual a 10")
else:
    print("x é menor ou igual a 5")
````

**Agora, vamos fazer um programa que classifica a nota de um aluno.**

````python
nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    print("Aluno aprovado!")
elif nota >= 5:
    print("Aluno em recuperação.")
else:
    print("Aluno reprovado.")
````

**Explicação:**

* Se a nota for maior ou igual a 7, o aluno é "aprovado".
* Se a nota for entre 5 e 7, o aluno está "em recuperação".
* Caso contrário, o aluno foi "reprovado"

# **5. Estruturas de Decisão Aninhadas**

Um if aninhado é quando temos um if dentro de outro if. Isso é útil para verificar condições mais complexas, onde a verificação de uma condição depende do resultado de outra.

**Sintaxe:**


```
if condição1:
    if condição2:
        # código para quando ambas as condições forem verdadeiras
    else:
        # código para quando a condição1 for verdadeira e a condição2 for falsa
else:
    # código para quando a condição1 for falsa
```



````python
nota = float(input("Digite a nota do aluno: "))
presenca = int(input("Digite a presença do aluno (em %): "))

if nota >= 7:
    if presenca >= 75:
        print("Aluno aprovado com boa nota e presença!")
    else:
        print("Aluno aprovado, mas com presença abaixo do necessário.")
else:
    if presenca >= 75:
        print("Aluno reprovado pela nota, mas com presença boa.")
    else:
        print("Aluno reprovado por nota e presença.")

````

**Explicação:**

* Se o aluno tiver uma nota maior ou igual a 7:
* Se a presença for maior ou igual a 75%, o aluno é aprovado com boa nota e presença.
* Caso contrário, o aluno é aprovado, mas com presença abaixo do necessário.
* Se a nota for abaixo de 7:
* Se a presença for maior ou igual a 75%, o aluno é reprovado por nota, mas com boa presença.
*Caso contrário, o aluno é reprovado tanto por nota quanto por presença.

# **6. Exemplos Mais Elaborados**


Vamos criar um sistema de votação, onde um usuário pode votar se tiver idade apropriada, e a votação pode ser para diferentes tipos de cargos (presidente, governador, etc.).

````python
idade = int(input("Digite sua idade: "))

if idade >= 16:
    cargo = input("Para qual cargo você deseja votar? (presidente, governador, prefeito): ").lower()
    if cargo == "presidente":
        print("Você pode votar para presidente.")
    elif cargo == "governador":
        print("Você pode votar para governador.")
    elif cargo == "prefeito":
        print("Você pode votar para prefeito.")
    else:
        print("Cargo inválido.")
else:
    print("Você não tem idade suficiente para votar.")
````

**Explicação:**

* Se o usuário tem 16 anos ou mais, ele pode escolher para qual cargo deseja votar.
* O programa verifica o cargo escolhido (presidente, governador, prefeito) e responde adequadamente.
* Caso a idade seja menor que 16, o programa avisa que o usuário não pode votar.

**Exemplo 2: Calculadora de Impostos**

Agora, vamos criar uma calculadora de impostos com base na faixa de renda do usuário.

````python
renda = float(input("Digite sua renda mensal: "))

if renda <= 2000:
    print("Você não paga imposto.")
elif renda <= 5000:
    imposto = renda * 0.10
    print(f"Você paga 10% de imposto, o valor é: R${imposto:.2f}")
elif renda <= 10000:
    imposto = renda * 0.15
    print(f"Você paga 15% de imposto, o valor é: R${imposto:.2f}")
else:
    imposto = renda * 0.20
    print(f"Você paga 20% de imposto, o valor é: R${imposto:.2f}")
````

**Explicação:**

* O programa verifica a faixa de renda do usuário e calcula o imposto correspondente:
* Até 2000: não paga imposto.
* De 2000 até 5000: paga 10%.
* De 5000 até 10000: paga 15%.
* Acima de 10000: paga 20%.
